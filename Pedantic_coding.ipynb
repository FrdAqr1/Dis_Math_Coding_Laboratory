### Given English Statement: 

<center> <b> Add additional fries and drinks to the order if the customer would buy a Cheese Burger Deluxe and add a Sundae instead if they also do. Always remember to print the receipt for every transaction. <center> <b>

<center> <i> p: Add additional fries and drinks 



<center> <i> q: the customer would buy a cheese burger deluxe 


<center> <i> r: Add sundae 


WFF: 

$$(p\rightarrow q)  \land (\neg p \rightarrow r) $$

Flowchart:

Code as:

```python
buy_cb = True
if(buy_cb):
    print(" Cheese burger deluxe w/ fries and drinks")
    print(" Orders complete")
else:
    print("Cheese burger deluxe w/ sundae")
    print(" Orders complete")
```

Code Explanation: 

 If it is true, the customer will buy a cheese burger deluxe, it will automatically add the additional fries and drinks followed by a receipt. Otherwise, if it is false, the customer will buy a cheese burger deluxe and will a add sundae as the substitute for the fries and drinks then print a receiept.
